```python
import pandas as pd
import ast

results_df = pd.read_csv("model_test_results.csv")
results_df['cm']
```

```python
# 將 cm 轉換為 numpy 陣列
results_df['cm'] = results_df['cm'].apply(lambda x: ast.literal_eval(x))
results_df.head()
```

```python
results_df['npv'] = results_df['cm'].apply(lambda x: x[0][0] / (x[0][0] + x[1][0]+0.0001))
results_df['recall'] = results_df['cm'].apply(lambda x: x[1][1] / (x[1][1] + x[1][0]+0.0001))
results_df['precision'] = results_df['cm'].apply(lambda x: x[1][1] / (x[1][1] + x[0][1]+0.0001))
results_df['f1'] = results_df['cm'].apply(lambda x: 2 * x[1][1] / (2 * x[1][1] + x[0][1] + x[1][0]+0.0001))
```

```python
results_df = results_df[results_df['precision'] != 0].reset_index(drop=True)

ratio = 0.5
results_df['score'] = ratio * results_df['npv'] + (1-ratio) * results_df['recall']
results_df.sort_values(by='score', ascending=False, ignore_index=True)
```
